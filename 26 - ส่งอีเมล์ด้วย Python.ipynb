## 2) `import` `library` มาบน Jupyter Notebook

```python
import yagmail
```

## 3) Login email ด้วย `.SMTP()`

สร้าง object ที่ใช้ส่ง email โดยการเข้าอีเมล์ของคุณ

```python
yag = yagmail.SMTP("yyyyyyyyy@gmail.com","xxxxxxxxx")
```

## 4) กำหนดผู้รับเป็น `string`

กำหนด email คนรับ

```python
receiver = "zzzzzzzzzzzzzzz@gmail.com"
```

## 5) กำหนด subject เป็น `string`

กำหนด subject

```python
subject = "Yagmail test with attachment"
```

## 6) เขียนเนื้อหาจดหมายเป็น `string`

กำหนดข้อความในจดหมาย

```python
body = "another one"
```

## 7) แนบรูปด้วย `.inline()`

แนบรูปเข้าไปในข้อความในจดหมาย

```python
body_pic = yagmail.inline("if.png")
```

## 8) ส่งอีเมล์ด้วย `.send()`

ส่ง email โดยใช้รายละเอียดดังกล่าว

```python
yag.send(
    to=receiver,
    subject=subject,
    contents=[body,body_pic], 
    attachments="if.png",
)
```
